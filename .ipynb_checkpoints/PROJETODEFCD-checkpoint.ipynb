**8- Plotar Intervalo de Confiança via Bootstrap do Número de Acidentes por Mês:
Utilize a técnica de bootstrap para calcular intervalos de confiança para o número de acidentes em cada mês. Plote esses intervalos juntamente com a média mensal dos acidentes para visualizar a variabilidade e a confiança das estimativas.**

```python
np.random.seed(43)

n_reamostra = 10000
media_boot= np.array([])

for i in range (n_reamostra):
    resample = anomes.sample(84, replace =True)
    media = resample.mean()
    media_boot = np.append(media_boot, media)

media_populacional = anomes.mean()

left = np.percentile(media_boot, 95)
right = np.percentile(media_boot, 5)

bpd.DataFrame().assign(bootstrap=media_boot).plot(kind='hist', density=True, bins=25, ec='w', figsize=(10, 5), color = 'purple')
plt.plot([left, right], [0, 0], color='gold', linewidth=12, label='Intervalo de confiança 90%', zorder=2);
plt.scatter(media_populacional, 0.000004, color='blue', s=100, label='Média da amostra').set_zorder(2)
plt.legend()
plt.ylabel("Frequência");
```

**10 - Use o Número de Acidentes por bairro de 2019 para Prever 2022:
Também use os de 2020
Por fim use os de 2022
Aplique métodos de previsão (como modelos de regressão) para estimar o número de acidentes em 2023, utilizando os dados passados.
Avalie os erros do modelo e discuta as limitações.
Compare os três modelos. A pandemia teve algum impacto, qual?**

```python
def padronizador(df):
    standardized_df = (df - df.mean()) / df.std()
    
    return padronizador
def unidades(df):
    media1=df.get('0_x').mean()
    media2=df.get('0_y').mean()
    std1 = np.std(df.get('0_x'))
    std2 = np.std(df.get('0_y'))
    pre_su1 = (df.get('0_x') - media1)/std1
    pre_su2 = (df.get('0_y') - media2)/std2

    su = bpd.DataFrame().assign(su1 = pre_su1, su2=pre_su2)
    return su 

def padrao(any_numbers):
    return (any_numbers - any_numbers.mean()) / np.std(any_numbers)

def correlation(df, x, y):
    x_su = padrao(df.get(x))
    y_su = padrao(df.get(y))
    return (x_su * y_su).mean()

def plot_data_and_line(dataset, x, y, point_0, point_1):
    dataset.plot(kind='scatter', x=x, y=y, label='data', figsize=(10, 5))
    xs, ys = zip(point_0, point_1)
    plt.plot(xs, ys, label='linha de regressão', color='orange')
    plt.xlabel('ano 1')
    plt.ylabel('ano 2')
    plt.legend()

def slope(df, x, y):
    "Returns the slope of the regression line between columns x and y in df (in original units)."
    r = correlation(df, x, y)
    return r * np.std(df.get(y)) / np.std(df.get(x))

def intercept(df, x, y):
    "Returns the intercept of the regression line between columns x and y in df (in original units)."
    return df.get(y).mean() - slope(df, x, y) * df.get(x).mean()
```

```python
## PREVENDO 2022 COM BASE EM 2019
su2019_2022 = unidades(dados_2019_2022)
r_19_22 = correlation(su2019_2022, 'su1', 'su2')
slope19_22 = slope(dados_2019_2022, '0_x', '0_y')
intercept19_22 = intercept(dados_2019_2022, '0_x', '0_y')
predict19_2022 = dados_2019_2022.get('0_x') * slope19_22 + intercept19_22
prevendo19_22 = bpd.DataFrame().assign(Acidentes2019 = dados_2019_2022.get('0_x'), acidentes2022 = dados_2019_2022.get('0_y'), previsao19_22 = predict19_2022 )

## PREVENDO 2022 COM BASE EM 2020
su2020_2022 = unidades(dados_2020_2022)
r_20_22 = correlation(su2020_2022, 'su1', 'su2')
slope20_22 = slope(dados_2020_2022, '0_x', '0_y')
intercept20_22 = intercept(dados_2020_2022, '0_x', '0_y')
predict20_2022 = dados_2020_2022.get('0_x') * slope20_22 + intercept20_22
prevendo20_22 = bpd.DataFrame().assign(Acidentes2020 = dados_2020_2022.get('0_x'), acidentes2022 = dados_2020_2022.get('0_y'), previsao20_22 = predict20_2022 )


## PREVENDO 2022 COM BASE EM 2021
su2021_2022 = unidades(dados_2021_2022)
r_21_22 = correlation(su2021_2022, 'su1', 'su2')
slope21_22 = slope(dados_2021_2022, '0_x', '0_y')
intercept21_22 = intercept(dados_2021_2022, '0_x', '0_y')
predict21_2022 = dados_2021_2022.get('0_x') * slope21_22 + intercept21_22
prevendo21_22 = bpd.DataFrame().assign(Acidentes2021 = dados_2021_2022.get('0_x'), acidentes2022 = dados_2021_2022.get('0_y'), previsao21_22 = predict21_2022 )

residuo19_22 = prevendo19_22.assign(residual = prevendo19_22.get('acidentes2022') - prevendo19_22.get('previsao19_22'))
residuo20_22 = prevendo20_22.assign(residual = prevendo20_22.get('acidentes2022') - prevendo20_22.get('previsao20_22'))
residuo21_22 = prevendo21_22.assign(residual = prevendo21_22.get('acidentes2022') - prevendo21_22.get('previsao21_22'))

print(residuo19_22.plot(kind='scatter', x='acidentes2022', xlabel='Numero de acidentes por bairro - 2022', y= 'residual', ylabel='Resíduos com base em 2019', color='purple'));
print(residuo20_22.plot(kind='scatter', x='acidentes2022', xlabel='Numero de acidentes por bairro - 2022', y= 'residual', ylabel='Resíduos com base em 2020', color='red'));
print(residuo19_22.plot(kind='scatter', x='acidentes2022', xlabel='Numero de acidentes por bairro - 2022', y= 'residual', ylabel='Resíduos com base em 2021', color='orange'));
```

*PREVISÕES PARA 2023*

```python
## Previsão para 2023 com base em 2019 e 2022
prever19_22_23= slope19_22 * dados_2019_2022.get('0_y') + intercept19_22
previsao19_22_23 = dados_2019_2022.assign(Previsao2023 = prever19_22_23)

## Previsão para 2023 com base em 2020 e 2022
prever20_22_23= slope20_22 * dados_2020_2022.get('0_y') + intercept20_22
previsao20_22_23 = dados_2020_2022.assign(Previsao2023 = prever20_22_23)

## Previsão para 2023 com base em 2021 e 2022
prever21_22_23= slope21_22 * dados_2021_2022.get('0_y') + intercept21_22
previsao21_22_23 = dados_2021_2022.assign(Previsao2023 = prever21_22_23)

## media dos acidentes por bairro de cada previsao
media23_19 = previsao19_22_23.get('Previsao2023').mean()
media23_20 = previsao20_22_23.get('Previsao2023').mean()
media23_21 = previsao21_22_23.get('Previsao2023').mean()
print(f'Previsão da média de de acidentes em 2023 usando 2019 como base 2019: {media23_19}')
print(f'Previsão da média de de acidentes em 2023 usando 2020 como base 2020: {media23_20}')
print(f'Previsão da média de de acidentes em 2023 usando 2021 como base 2021: {media23_21}')
```
